```python
def repartition_parquet(taxi_type = 'green', year = 2020, repartitions = 10):
    for month in range(1,13):
        input_path = f"../datasets/raw/{taxi_type}/{year}/{taxi_type}_tripdata_{year}-{month:02d}.parquet" # 02d means pad with 0, string width = 2 (by default, right aligned), d means decimal format (integer)
        output_path = f"../datasets/pq/{taxi_type}/{year}/{month:02d}/" # Provide folder and not file or else partitioning doesnt work

        # Reading parquet files
        df_taxi = spark.read \
            .parquet(input_path)

        # Skipping Schema enforcement since it doesnt need one

        # Performing repartition for parallelism
        df_taxi \
            .repartition(repartitions) \
            .write \
            .parquet(output_path)
```

```python
repartition_parquet('green',2020)
repartition_parquet('green',2021)
repartition_parquet('yellow',2020)
repartition_parquet('yellow',2021)
```
